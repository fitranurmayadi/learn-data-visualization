##Pandas Intermediate: indexing and Selection

###Select Column

Method 1: quick and easy, but have limitations

```python
    df.column_name               # returns series
```

Method 2: more robust and can filter on multiple columns
```python
    df['column']                  # returns series
```

```python
    df[['column 1', 'column 2']]  # returns dataframe
```

Example

````python
df_can.Country
````

filtering on list of countries and data for years

````python
df_can[['Country', 1980, 1981, 1982,1983,1984,1985]]
````

###Select Row

```python
    df.loc[label]    # filters by the labels of the index/column
    df.iloc[index]   # filters by the positions of the index/column
```

###Plotting in Pandas

##Line Pots (Series/Dataframe)

Example

````python
haiti = df_can.loc['Haiti', years]
haiti.head()
````

use .plot() function to the haiti dataframe

````python
haiti.plot()
````

Add label in x and y axis

````python
haiti.index = haiti.index.map(int)
haiti.plot(kind='line')

plt.title('Immigrant from Haiti')
plt.ylabel('Numbers of immigrants')
plt.xlabel('Years')
plt.show()
````

````python
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

# annotate the 2010 Earthquake. 
plt.text(2000, 6000, '2010 Earthquake')
plt.show() 
````

Example for China and India data

````python
df_CI = df_can.loc[['India','China'], years]
````

````python
df_CI = df_CI.transpose()
df_CI.head()
````

````python
df_CI.index = df_CI.index.map(int) 
df_CI.plot(kind='line')

plt.title('Immigrants from China and India')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
````

Example: top 5 most immigrant countries

Step 1: getdataset, sort values to top 5 countries

````python
df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)
df_top5 = df_can.head(5)

df_top5 = df_top5[years].transpose()

print(df_top5.columns)
````

Step 2: Plot dataframe

````python
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='line', figsize=(19,7))

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immmigrants')
plt.xlabel('Years')
plt.show()
````
